# Shoe Sell-Through — Exploratory Data Analysis

**Goal:** understand the target variable distribution, validate features, and establish tier-label thresholds before model training.

**Constraints enforced throughout:** only signals available *at time of receipt* are used as features — no sales velocity, markdown rate, or any post-receipt data.

```python
import sqlite3
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.ticker as mtick
import seaborn as sns
from pathlib import Path

sns.set_theme(style="whitegrid", palette="muted")
plt.rcParams["figure.dpi"] = 110

DB = Path("../sellthrough.db")
```

```python
conn = sqlite3.connect(DB)
df = pd.read_sql_query("SELECT * FROM sku_sell_through", conn)
conn.close()

df["receive_month"] = pd.to_datetime(df["receive_date"]).dt.month

print(f"SKUs: {len(df)}")
print(f"Columns: {list(df.columns)}")
df.head()
```

---
## 1. Target Variable: Sell-Through Rate Distribution

Sell-through rate = `units_sold / units_received` per SKU, computed over the full data window.

```python
fig, axes = plt.subplots(1, 2, figsize=(13, 4))

# Histogram
axes[0].hist(df["sell_through_rate"], bins=20, edgecolor="k", color="steelblue")
axes[0].axvline(df["sell_through_rate"].median(), color="red", linestyle="--",
                label=f"Median: {df['sell_through_rate'].median():.2f}")
axes[0].set_xlabel("Sell-Through Rate")
axes[0].set_ylabel("# SKUs")
axes[0].set_title("Sell-Through Rate Distribution")
axes[0].xaxis.set_major_formatter(mtick.PercentFormatter(1.0))
axes[0].legend()

# Sorted bar — each SKU
sorted_rates = df["sell_through_rate"].sort_values(ascending=False).reset_index(drop=True)
axes[1].bar(range(len(sorted_rates)), sorted_rates, color="steelblue", edgecolor="none")
axes[1].set_xlabel("SKU rank")
axes[1].set_ylabel("Sell-Through Rate")
axes[1].set_title("Sell-Through Rate by SKU (ranked)")
axes[1].yaxis.set_major_formatter(mtick.PercentFormatter(1.0))

plt.tight_layout()
plt.show()

df["sell_through_rate"].describe().apply(lambda x: f"{x:.3f}")
```

---
## 2. Tier Labels

Thresholds are aligned with the data-generation tier ranges (see `generate_data.py`). Adjust if the distribution above suggests different natural breaks.

```python
# These thresholds will be reused verbatim in model training.
TIER_THRESHOLDS = {
    "hot":      0.75,   # >= 0.75
    "moderate": 0.45,   # 0.45 – 0.75
    "slow":     0.25,   # 0.25 – 0.45
    # dead:     < 0.25
}
TIER_ORDER = ["hot", "moderate", "slow", "dead"]
TIER_COLORS = {"hot": "#2ca02c", "moderate": "#1f77b4", "slow": "#ff7f0e", "dead": "#d62728"}

def assign_tier(rate: float) -> str:
    if rate >= TIER_THRESHOLDS["hot"]:
        return "hot"
    if rate >= TIER_THRESHOLDS["moderate"]:
        return "moderate"
    if rate >= TIER_THRESHOLDS["slow"]:
        return "slow"
    return "dead"

df["tier"] = df["sell_through_rate"].apply(assign_tier)

tier_counts = df["tier"].value_counts().reindex(TIER_ORDER).fillna(0).astype(int)
print(tier_counts.to_string())

fig, ax = plt.subplots(figsize=(6, 4))
tier_counts.plot.bar(ax=ax, color=[TIER_COLORS[t] for t in TIER_ORDER], edgecolor="k", width=0.6)
ax.set_title("Tier Distribution (n=" + str(len(df)) + " SKUs)")
ax.set_xlabel("")
ax.set_ylabel("# SKUs")
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

---
## 3. Price Tier

Buckets: budget (<$80), mid ($80–$130), premium ($130–$175), luxury ($175+). Computed from `avg_retail_price` in the `sku_sell_through` view.

```python
PRICE_ORDER = ["budget", "mid", "premium", "luxury"]
pt_present = [p for p in PRICE_ORDER if p in df["price_tier"].values]

fig, axes = plt.subplots(1, 2, figsize=(13, 4))

# Box: sell-through by price tier
df.boxplot(column="sell_through_rate", by="price_tier",
           positions=range(len(pt_present)),
           ax=axes[0])
axes[0].set_xticklabels(pt_present, rotation=0)
axes[0].yaxis.set_major_formatter(mtick.PercentFormatter(1.0))
axes[0].set_title("Sell-Through Rate by Price Tier")
axes[0].set_xlabel("")
plt.sca(axes[0])
plt.title("Sell-Through Rate by Price Tier")

# Stacked bar: tier composition within each price tier
cross = (
    pd.crosstab(df["price_tier"], df["tier"])
      .reindex(index=pt_present, columns=TIER_ORDER, fill_value=0)
)
cross.plot.bar(ax=axes[1], stacked=True,
               color=[TIER_COLORS[t] for t in TIER_ORDER],
               edgecolor="k", width=0.6)
axes[1].set_title("Tier Composition by Price Tier")
axes[1].set_xlabel("")
axes[1].set_ylabel("# SKUs")
plt.xticks(rotation=0)
axes[1].legend(title="tier", bbox_to_anchor=(1.02, 1), loc="upper left")

plt.tight_layout()
plt.show()

print(df.groupby("price_tier")["sell_through_rate"].agg(["count", "mean", "median"])
        .reindex(pt_present)
        .round(3)
        .to_string())
```

---
## 4. Department

```python
dept_order = (
    df.groupby("department")["sell_through_rate"]
      .median()
      .sort_values(ascending=False)
      .index.tolist()
)

fig, ax = plt.subplots(figsize=(9, 4))
df.boxplot(column="sell_through_rate", by="department",
           positions=range(len(dept_order)),
           ax=ax)
ax.set_xticklabels(dept_order, rotation=15, ha="right")
ax.yaxis.set_major_formatter(mtick.PercentFormatter(1.0))
ax.set_title("Sell-Through Rate by Department")
ax.set_xlabel("")
plt.title("Sell-Through Rate by Department")
plt.tight_layout()
plt.show()

print(df.groupby("department")["sell_through_rate"]
        .agg(["count", "mean", "median"])
        .reindex(dept_order)
        .round(3)
        .to_string())
```

---
## 5. Vendor

```python
vendor_order = (
    df.groupby("vendor")["sell_through_rate"]
      .median()
      .sort_values(ascending=False)
      .index.tolist()
)

fig, ax = plt.subplots(figsize=(9, 4))
df.boxplot(column="sell_through_rate", by="vendor",
           positions=range(len(vendor_order)),
           ax=ax)
ax.set_xticklabels(vendor_order, rotation=0)
ax.yaxis.set_major_formatter(mtick.PercentFormatter(1.0))
ax.set_title("Sell-Through Rate by Vendor")
ax.set_xlabel("")
plt.title("Sell-Through Rate by Vendor")
plt.tight_layout()
plt.show()

print(df.groupby("vendor")["sell_through_rate"]
        .agg(["count", "mean", "median"])
        .reindex(vendor_order)
        .round(3)
        .to_string())
```

---
## 6. Seasonality: Receive Month

Does the month a shoe arrives predict how well it sells through? Earlier receipts have more selling time in the data window; later receipts are disadvantaged.

```python
month_labels = {1: "Jan", 2: "Feb", 3: "Mar", 4: "Apr", 5: "May", 6: "Jun"}
df["receive_month_label"] = df["receive_month"].map(month_labels)

month_order = [month_labels[m] for m in sorted(df["receive_month"].unique())]

fig, axes = plt.subplots(1, 2, figsize=(13, 4))

df.groupby("receive_month_label")["sell_through_rate"].mean().reindex(month_order).plot.bar(
    ax=axes[0], color="steelblue", edgecolor="k", width=0.6
)
axes[0].yaxis.set_major_formatter(mtick.PercentFormatter(1.0))
axes[0].set_title("Mean Sell-Through Rate by Receive Month")
axes[0].set_xlabel("")
plt.sca(axes[0])
plt.xticks(rotation=0)

df.groupby("receive_month_label")["sku"].count().reindex(month_order).plot.bar(
    ax=axes[1], color="grey", edgecolor="k", width=0.6
)
axes[1].set_title("# SKUs Received per Month")
axes[1].set_xlabel("")
axes[1].set_ylabel("# SKUs")
plt.sca(axes[1])
plt.xticks(rotation=0)

plt.tight_layout()
plt.show()
```

---
## 7. Buy Quantity (Units Received)

A larger initial buy may signal buyer confidence — or create harder-to-sell excess. Check for correlation with sell-through rate.

```python
fig, ax = plt.subplots(figsize=(7, 4))

for tier in TIER_ORDER:
    sub = df[df["tier"] == tier]
    ax.scatter(sub["units_received"], sub["sell_through_rate"],
               label=tier, color=TIER_COLORS[tier], alpha=0.75, edgecolors="k", linewidths=0.4)

ax.set_xlabel("Units Received (style total)")
ax.set_ylabel("Sell-Through Rate")
ax.yaxis.set_major_formatter(mtick.PercentFormatter(1.0))
ax.set_title("Units Received vs. Sell-Through Rate")
ax.legend(title="tier")
plt.tight_layout()
plt.show()

corr = df[["units_received", "avg_retail_price", "receive_month", "sell_through_rate"]].corr()
print(f"Correlation with sell_through_rate:\n{corr['sell_through_rate'].drop('sell_through_rate').round(3).to_string()}")
```

---
## 8. Feature Correlation Matrix

Encode categoricals numerically to inspect pairwise correlations with the target.

```python
encode_cols = ["vendor", "department", "price_tier"]
df_enc = df[["sell_through_rate", "avg_retail_price", "units_received", "receive_month"] + encode_cols].copy()

for col in encode_cols:
    df_enc[col] = df_enc[col].astype("category").cat.codes

corr = df_enc.corr()

fig, ax = plt.subplots(figsize=(8, 6))
mask = corr.isnull()
sns.heatmap(corr, annot=True, fmt=".2f", cmap="coolwarm", center=0,
            linewidths=0.5, ax=ax, mask=mask)
ax.set_title("Feature Correlation Matrix")
plt.tight_layout()
plt.show()
```

---
## 9. Class Balance & Modeling Notes

```python
print("=== Class distribution ===")
print(df["tier"].value_counts().reindex(TIER_ORDER).to_string())
print()

n = len(df)
min_class = df["tier"].value_counts().min()

print(f"Total SKUs      : {n}")
print(f"Smallest class  : {min_class} samples")
print()
print("Modeling notes:")
print(f"  - Dataset is small ({n} SKUs). Use stratified k-fold CV (k=5) or leave-one-out if any class < 10 samples.")
print("  - Prefer low-variance models: logistic regression, random forest (max_depth ≤ 3), gradient boosting (n_estimators ≤ 100).")
print("  - If classes are imbalanced, use class_weight='balanced' or stratified resampling.")
print("  - Features available at receipt: vendor, department, price_tier, avg_retail_price, receive_month, units_received.")
print("  - Target: 4-class tier label derived from sell_through_rate using TIER_THRESHOLDS defined above.")
```
